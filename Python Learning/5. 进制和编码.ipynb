# 4. 编码
编码，文字和二进制之间的对照表

## 4.1 ascii编码
ascii规定使用1个字节来表示字母与二进制的对应关系。  
00000000 2**8=256

## 4.2 gb-2312编码
gb-2312：国家信息标准委员会
gbk：对gb-2312的扩展，包含中日韩等文字

在二进制对应关系时，由如下逻辑：
* 单字节表示，用一个字节表示对应关系
* 双字节表示，用两个字节表示对应关系

## 4.3 unicode
unicode也被称为万国码，为全球的每个文字都分配了一个码位（二进制表示）
* ucs2：用固定的两个字节去表示一个文字
      2**16 65535
* ucs4：用固定的四个字节表示一个文字
      2**32

  unicode的应用：在文件存储和网络传输时，不会直接使用unicode。而在内存中会用unicode，固定长度便于计算，而计算量不大
## utf-8
包含所有文字和二进制的对饮关系，全球应用量最为广泛的编码。  
本质上：是对unicode的压缩，用尽量少的二进制去对应文字。  
具体的压缩过程：  
* 第一步：选择模板
* 第二步：在模板中填入数据
    - 转换为二进制
    - 根据模板套数据
