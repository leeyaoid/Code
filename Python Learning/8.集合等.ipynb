# 2. 字典
字典是 有序、键不重复 且 元素只能是键值对的可变的 容器。
* 容器
* 元素必须键值对
* 键不重复，重复则会被覆盖
## 2.1 定义
字典中对键值对的要求：
* 键：必须可哈希。int/bool/str/tuple          不可哈希：lsit/set/dict
* 值：任意类型
