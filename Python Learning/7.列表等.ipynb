# 1. 列表（list）
列表（list），是一个**有序且可变**的容器，在里面可以存放多个不同类型的元素。
## 1.1 定义
不可变类型：字符串、布尔、整型
可变类型：列表（内部数据可以修改）

## 1.2 独有功能

### 1. 追加，在原列表尾部追加值
.append()  
  

### 2.批量追加，将一个列表追加到另一个列表
.extend()  


### 3.插入，在原列表指定索引插入值
.insert(index,"xxx")  


### 4.删除（慎用，里面没有会报错）


### 5.根据索引位置删除
.pop()  
### 6.清空
.clear()
### 7. 根据值找到索引位置（慎用，找不到报错）
.index()
## 8. 列表元素排序
.sort()

“我儿子永远不能换成别人，但是我儿子可以长大”（元组之中有列表）
## 2.1 定义


```python
v1 = ("alex", "b",) #建议元组再最后多加一个逗号
v2 =  ("alex") #不是元组
v2 =  ("alex", ) #元组
```
